```python
import matplotlib.pyplot as plt
import seaborn as sb
import pandas as pd
import geopandas as gpd
import numpy as np
```

# ETL

```python
df_pop=pd.read_csv('https://raw.githubusercontent.com/awdippolito/Palette_Assignments/main/Merged%20Country%20Population%20Data.csv')
```

```python
df_wh=pd.read_csv('https://raw.githubusercontent.com/ruhidm/Palette_Cohort_3_Data/main/World_countries_wheat_production_cleaned.csv')
```

```python
df_con=pd.read_html('https://statisticstimes.com/geography/countries-by-continents.php')[2]\
    [['Country or Area','Continent']]\
        .rename(columns= {'Country or Area': 'Country'})
```

```python
gdf=gpd.read_file('World_Countries_Generalized.geojson')
```

```python
df_pop
```

```python
df_wh
```

```python
pd.merge(df_wh, df_pop, on='Country', how='inner')
```

```python
df_pop['Country'].unique()[:180]
```

```python
df_wh['Country'].unique()
```

```python
df_wh['Country']=df_wh['Country'].replace(to_replace={'Czechia': 'Czech Republic'})
```

```python
df=pd.merge(df_wh, df_pop, on='Country', how='inner')
```

```python
df['PerPerson']=df['Average'] / df['Population']
```

```python
df=df.set_index('Country')
```

```python
df.head()
```

```python
df[['Average', 'Population', 'PerPerson']].sort_values('Average', ascending=False)['Average']\
    .plot(kind='bar', color='red')
plt.title('Average Wheat Production', size=18, color='red')
plt.xlabel('Countries', size=14, color='red')
plt.ylabel('Wheat in kgs (x1000 000)', size=14, color='red')
plt.xticks(size=10, color='gray')
plt.yticks(size=10, color='gray')
plt.axhline(np.mean(df['Average']), color='green',linestyle='--')
plt.tight_layout()
plt.show()
```

```python
gpd.GeoDataFrame(pd.merge(
    df.reset_index()[['Country', 'Average','Population', 'PerPerson']], 
    gdf.rename(columns={'COUNTRY': 'Country'})[['Country', 'geometry']], 
    on='Country', 
    how='inner')).plot(column='Average', cmap='rainbow', legend=True) # Use .explore() for interactive map
```

```python
gpd.GeoDataFrame(pd.merge(
    df.reset_index()[['Country', 'Average','Population', 'PerPerson']], 
    gdf.rename(columns={'COUNTRY': 'Country'})[['Country', 'geometry']], 
    on='Country', 
    how='inner')).plot(column='PerPerson', cmap='rainbow', legend=True) # Use .explore() for interactive map
```

```python
df_con['Country']=df_con['Country'].replace(to_replace={'Czechia': 'Czech Republic',
                                                         'Russian Federation': 'Russia',
                                                         'Iran (Islamic Republic of)': 'Iran',
                                                         'United Kingdom of Great Britain and Northern Ireland': 'United Kingdom',
                                                         'Syrian Arab Republic': 'Syria',
                                                         'United States of America': 'United States'})
```

```python
gdf = gdf.rename(columns={'COUNTRY': 'Country'})[['Country', 'geometry']]
```

```python
gdf['Country'] = gdf['Country'].replace(to_replace= {'Russian Federation': 'Russia',
                                   'Turkiye': 'Turkey'})
```

```python
gpd.GeoDataFrame(pd.merge(
    df.reset_index()[['Country', 'Average','Population', 'PerPerson']], 
    gdf, 
    on='Country', 
    how='inner'))
```

```python
gpd.GeoDataFrame(
pd.merge(
    # First dataframe
    gpd.GeoDataFrame(pd.merge(
    df.reset_index()[['Country', 'Average','Population', 'PerPerson']], 
    gdf.rename(columns={'COUNTRY': 'Country'})[['Country', 'geometry']], 
    on='Country', 
    how='inner')),

    #Second Dataframe
    df_con,
    on='Country',
    how='inner')).to_file('World_Wheat_Ind.geojson', driver='GeoJSON') # you can save as shapely file - replace geojon with .shp
```
